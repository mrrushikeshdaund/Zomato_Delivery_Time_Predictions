## Zomato Delivery Time Predictions

##### Introduction About the Data : The dataset The goal is to predict Delivery time of given Zomato (Regression Analysis).


```python
import pandas as pd
```

```python
df = pd.read_csv("data/Zomato_Delivery_Dataset.csv")
df.head()
```

```python
df = df.drop(labels=['ID','Delivery_person_ID'],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Delivery_person_Age'] = df['Delivery_person_Age'].fillna(value=df['Delivery_person_Age'].mean())
```

```python
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].fillna(value=str(df['Delivery_person_Ratings'].mode()))
```

```python
df['Time_Orderd'] = df['Time_Orderd'].fillna(value=str(df['Time_Orderd'].mode()))
```

```python
df['Weather_conditions'] = df['Weather_conditions'].fillna(value=str(df['Weather_conditions'].mode()))
```

```python
df['Road_traffic_density'] = df['Road_traffic_density'].fillna(value=str(df['Road_traffic_density'].mode()))
```

```python
df['multiple_deliveries'] = df['multiple_deliveries'].fillna(value=int(0))
```

```python
df['Festival'] = df['Festival'].fillna(value=str(df['Festival'].mode()))
```

```python
df['City'] = df['City'].fillna(value=str(df['City'].mode()))
```
